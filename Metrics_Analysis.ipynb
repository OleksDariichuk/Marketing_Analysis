```python
def check_countries(df):
    # get a list of country abbreviations
    country_codes = [country.alpha_2 for country in list(pycountry.countries)]

    # Check if all values in the field 'Country' belong to the list of countries
    invalid_countries = df[~df['Country'].isin(country_codes)]['Country'].unique()

    if len(invalid_countries) > 0:
        print(f"Invalid values in the field 'Country': {invalid_countries}")
    else:
        print("All values in the 'Country' field are valid.")
```

```python
calculations = pd.merge(data1, data2, on='Country')


# CPM (Cost per mille)
calculations['CPM'] = (calculations['Amount spent'] / calculations['Impressions']) * 1000

# CTR (Click-through rate)
calculations['CTR'] = (calculations['Emails'] / calculations['Impressions']) * 100

# CPA (Cost per email acquisition)
calculations['CPA'] = calculations['Amount spent'] / calculations['Emails']

# CR into email enter (Conversion rate into email enter)
calculations['CR into email enter'] = (calculations['Emails'] / calculations['Visits']) * 100

# CR into paywall view (Conversion rate into paywall view)
calculations['CR into paywall view'] = (calculations['Paywalls'] / calculations['Visits']) * 100

# CR into purchase (Conversion rate into purchase)
calculations['CR into purchase'] = (calculations['Paywalls'] / calculations['Emails']) * 100
```

```python
not_merged_data1 = data1[~data1['Country'].isin(calculations['Country'])]
not_merged_data2 = data2[~data2['Country'].isin(calculations['Country'])]
```

```python
from datetime import datetime

transactions = data3[data3['Status'] == 'settled']
transactions = transactions[transactions['Payment type']=='Recurring'] #!

# Calculate lifespan for each user
user_lifespan = transactions.groupby('User ID')['Created at'].agg(['min', 'max'])
eq = ((user_lifespan['max'].eq(user_lifespan['min'])))
lifespan = []
for i, row in user_lifespan.iterrows():
    if row['max'] == row['min']:
        lifespan.append(1)
    else:
        lifespan.append((row['max'] - row['min'])/ np.timedelta64(1, 'M')+1)

user_lifespan['lifespan'] = lifespan


# Group transactions by country
country_groups = transactions.groupby('Country')

# Calculate the measured interval, avg transaction price, avg subscription, and avg lifetime for each country
country_metrics = []
for country, country_data in country_groups:
    
    avg_transaction_price = country_data['Amount'].sum() / country_data['Amount'].count()
    
    unique_users =  country_data['User ID'].nunique()
    
    avg_lifetime = user_lifespan[user_lifespan.index.isin(country_data['User ID'])]['lifespan'].mean()
    country_metrics.append({
        'Country': country,
        'Avg transaction price': avg_transaction_price,
        'unique users': unique_users,
        'Avg lifetime (months)': avg_lifetime
    })

# Convert country_metrics to a DataFrame and calculate the LTV for each country
country_metrics_df = pd.DataFrame(country_metrics)
country_metrics_df['LTV'] = country_metrics_df['Avg transaction price'] * (country_metrics_df['Avg lifetime (months)'])


print(country_metrics_df)
```

```python
calculations = pd.merge(calculations, country_metrics_df, on='Country')
```

```python
print(calculations)
```

```python
calculations['ROMI'] = (calculations['LTV']*calculations['unique users'] - calculations['Amount spent'])   / calculations['Amount spent'] 
```

```python
calculations.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
calculations = calculations.sort_values(by='Amount spent', ascending=False)
```
